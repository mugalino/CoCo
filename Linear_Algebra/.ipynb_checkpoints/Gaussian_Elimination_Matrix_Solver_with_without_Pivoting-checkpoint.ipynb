```python
# author: mugalino 2022

# Imports the necessary packages
import numpy as np
from time import time
import matplotlib.pyplot as plt

# This function implements Gaussian elimination with or without pivoting
def gaussian_elim(A, b, pivoting = False, debug = False, print_new_matrix = False, print_vector_solution = True, check_rms_error = False): 
    """
    This is a function that takes in an input matrix A and vector b, and 
    returns an upper triangular matrix to be used for a back
    substitution procedure to find the solution x for the matrix
    equation (Ax = b).
    
    mugalino 2022
    """
    
    start_time = time()
    
    # START OF GAUSSIAN ELIMINATION STEP - UPPER TRIANGULAR 
    # WORKS WITH OR WITHOUT PIVOTING, SEE OPTIONS
    
    mat = np.array(A)
    orig_b = b
    b = np.array(b)
    mat_shape = np.shape(A)[0]
    
    if debug:
        print("Original matrix A:\n", mat)
    for column in range(mat_shape-1):     
        if debug:
            print("Currently on Column:" + str(column+1))
            
        current_column = mat[:, column]
        
        # First check if we require partial pivoting
        if (pivoting and column != mat_shape-1):
            max_index = np.argmax(current_column[column:])
            if debug:
                print("Current Column:", current_column[column:])

                print("Swapping:\n", mat[[column, (max_index + column)]])
            
            mat[[column, (max_index + column)]] = mat[[(max_index + column), column]]
            b[[column, (max_index + column)]] = b[[(max_index + column), column]]
            
            # swaps row with max with diagonal row       
            current_column = mat[:, column]
            
            if debug:
                print("Pivoted matrix:\n", mat)
                print("New vector b:", b)
        
        # Computes factors that we need for turning column elements [column_index + 1 : ] to zero
        factors_m = np.array([current_column[row]/current_column[column] for row in range(column + 1, mat_shape)])
        
        # Computes new rows using computed factors
        new_mat = np.array([mat[row, :] - factors_m[row - (column + 1)] * mat[column, :] for row in range(column + 1, mat_shape)])
        
        # Updates original matrix 
        mat[column+1:, :] = new_mat
        
        # Updates solution vector
        b[column+1:] = b[column+1:] - factors_m * b[column] 
        
        if debug:
            print("New matrix after step %04d" % (column+1) + "\n", mat)
            
    if print_new_matrix:
        if pivoting:
            print("Matrix after Gaussian elimination with pivoting:\n", mat)
            print("Vector b after Gaussian elimination with pivoting:\n", b)
        else:
            print("Matrix after Gaussian elimination without pivoting:\n", mat)
            print("Vector b after Gaussian elimination without pivoting:\n", b)
        
    # END OF GAUSSIAN ELIMINATION
    
    # START OF BACK SUBSTITUTION STEP
    solution_vector = np.zeros(len(b))
    
    for index in range(len(solution_vector)-1,-1,-1):
        solution_vector[index] = (b[index] - np.sum(solution_vector[:]*mat[index,:]))/mat[index, index]
    
    end_time = time()
    
    # Check against well tested linear algebra packages on numpy
    if check_rms_error:
        numpy_solution = np.linalg.solve(A, orig_b)
        
        if debug:
            print("Numpy LAPACK solution:", numpy_solution)
        rms_error = np.sqrt(np.average((numpy_solution - solution_vector)**2))
        print("RMS Error against LAPACK Numpy:", rms_error)
        
        if debug:
            plt.plot(range(len(solution_vector)), numpy_solution)
    
    if print_vector_solution:
        if pivoting:
            print("Solution x of matrix equation Ax = b, with partial pivoting:", solution_vector)
        else:
            print("Solution x of matrix equation Ax = b, without partial pivoting:", solution_vector)
    print("Execution time:", end_time - start_time)
    
    plt.plot(range(len(solution_vector)), solution_vector)
    plt.title(r"Plot of solution vector as a function of index $\vec{x}(j)$")
    plt.xlabel("Index $j$")
    plt.ylabel(r"Vector $\vec{x}$")
```
